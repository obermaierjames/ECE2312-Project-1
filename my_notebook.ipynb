```python
#Import Necessary Libraries
import numpy as np
import matplotlib.pyplot as plt
import sounddevice as sd
from scipy.io.wavfile import write
import scipy.io.wavfile as wav
import math
```

```python
#Section 3 - Record Phrase 1
#You have to record your own phrases for the bottom sections to run properly

fs = 44100  # Sample rate
seconds = 5  # Duration of recording

Phrase1 = sd.rec(int(seconds * fs), samplerate=fs, channels=2)
sd.wait()  # Wait until recording is finished
```

```python
#Section 3 - Plot Phrase 1

#Convert 2 channel audio to 1 channel audio
mono_waveform1 = np.mean(Phrase1, axis=1)

#Convert from samples into time using the sampling rate; # samples/sampling rate = seconds
time = np.arange(0, len(Phrase1)) / fs

#Plot the waveform data over time
plt.figure(figsize=(10, 4))
plt.plot(time, mono_waveform1)
plt.title('Waveform of Phrase 1' )
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

```python
#Section 6 - Stereo effects using Phrase 1

#Read in phrase 1 wav file
fs, Phrase1 = wav.read('Phrase1.wav')

# Split stereo audio into left and right channels
left_channel = Phrase1[:, 0]
right_channel = Phrase1[:, 1]

#Delay by the average head - 0.5ms or 22 samples
Delay1 = math.floor(fs * 0.0005) #Calculates the number of samples corresponding to delay in seconds (rounds down to nearest int)
left_channel_delayed1 = np.pad(left_channel, (0, Delay1), mode='constant')
right_channel_delayed1 = np.pad(right_channel, (Delay1, 0), mode='constant')

#Delay by 1ms or 44 samples
Delay2 = math.floor(fs * 0.001) #Calculates the number of samples corresponding to delay in seconds (rounds down to nearest int)
left_channel_delayed2 = np.pad(left_channel, (0, Delay2), mode='constant')
right_channel_delayed2 = np.pad(right_channel, (Delay2, 0), mode='constant')

#Delay by 10ms or 441 samples
Delay3 = math.floor(fs * 0.01) #Calculates the number of samples corresponding to delay in seconds (rounds down to nearest int)
left_channel_delayed3 = np.pad(left_channel, (0, Delay3), mode='constant')
right_channel_delayed3 = np.pad(right_channel, (Delay3, 0), mode='constant')

#Delay by 100ms or 4410 samples
Delay4 = math.floor(fs * 0.1) #Calculates the number of samples corresponding to delay in seconds (rounds down to nearest int)
left_channel_delayed4 = np.pad(left_channel, (0, Delay4), mode='constant')
right_channel_delayed4 = np.pad(right_channel, (Delay4, 0), mode='constant')

#Recombine left and right channel for stereo audio
stereo_audio = np.column_stack((left_channel, right_channel))
stereo_audio_delay1 = np.column_stack((left_channel_delayed1, right_channel_delayed1))
stereo_audio_delay2 = np.column_stack((left_channel_delayed2, right_channel_delayed2))
stereo_audio_delay3 = np.column_stack((left_channel_delayed3, right_channel_delayed3))
stereo_audio_delay4 = np.column_stack((left_channel_delayed4, right_channel_delayed4))

#Write each stereo file above to its own wav file
write('Team[Obermaier]-stereosoundfile-[0].wav', fs, stereo_audio)
write('Team[Obermaier]-stereosoundfile-[avghead].wav', fs, stereo_audio_delay1)
write('Team[Obermaier]-stereosoundfile-[1].wav', fs, stereo_audio_delay2)
write('Team[Obermaier]-stereosoundfile-[10].wav', fs, stereo_audio_delay3)
write('Team[Obermaier]-stereosoundfile-[100].wav', fs, stereo_audio_delay4)
```

```python
#Section 6 - Stereo effects using Phrase 1

#Read in phrase 1 delayed
fs, StereoDelay1 = wav.read('Team[Obermaier]-stereosoundfile-[0].wav')
fs, StereoDelay2 = wav.read('Team[Obermaier]-stereosoundfile-[avghead].wav')
fs, StereoDelay3 = wav.read('Team[Obermaier]-stereosoundfile-[1].wav')
fs, StereoDelay4 = wav.read('Team[Obermaier]-stereosoundfile-[10].wav')
fs, StereoDelay5 = wav.read('Team[Obermaier]-stereosoundfile-[100].wav')

#Split right channels
right_channel1 = StereoDelay1[:, 1]
right_channel2 = StereoDelay2[:, 1]
right_channel3 = StereoDelay3[:, 1]
right_channel4 = StereoDelay4[:, 1]
right_channel5 = StereoDelay5[:, 1]

#Attenuate by 6dB
one_attenuate1 = right_channel1 * 0.5
one_attenuate2 = right_channel2 * 0.5
one_attenuate3 = right_channel3 * 0.5
one_attenuate4 = right_channel4 * 0.5
one_attenuate5 = right_channel5 * 0.5

#Attenuate by 3dB
two_attenuate1 = right_channel1 * 0.707
two_attenuate2 = right_channel2 * 0.707
two_attenuate3 = right_channel3 * 0.707
two_attenuate4 = right_channel4 * 0.707
two_attenuate5 = right_channel5 * 0.707

#Attenuate by 1.5dB
three_attenuate1 = right_channel1 * 0.841
three_attenuate2 = right_channel2 * 0.841
three_attenuate3 = right_channel3 * 0.841
three_attenuate4 = right_channel4 * 0.841
three_attenuate5 = right_channel5 * 0.841

#Recombine 6dB attenuated stereo
Delay1_6dB = np.column_stack((StereoDelay1[:, 0], one_attenuate1))
Delay2_6dB = np.column_stack((StereoDelay2[:, 0], one_attenuate2))
Delay3_6dB = np.column_stack((StereoDelay3[:, 0], one_attenuate3))
Delay4_6dB = np.column_stack((StereoDelay4[:, 0], one_attenuate4))
Delay5_6dB = np.column_stack((StereoDelay5[:, 0], one_attenuate5))

#Recombine 3dB attenuated stereo
Delay1_3dB = np.column_stack((StereoDelay1[:, 0], two_attenuate1))
Delay2_3dB = np.column_stack((StereoDelay2[:, 0], two_attenuate2))
Delay3_3dB = np.column_stack((StereoDelay3[:, 0], two_attenuate3))
Delay4_3dB = np.column_stack((StereoDelay4[:, 0], two_attenuate4))
Delay5_3dB = np.column_stack((StereoDelay5[:, 0], two_attenuate5))

#Recombine 1.5dB attenuated stereo
Delay1_1dB = np.column_stack((StereoDelay1[:, 0], three_attenuate1))
Delay2_1dB = np.column_stack((StereoDelay2[:, 0], three_attenuate2))
Delay3_1dB = np.column_stack((StereoDelay3[:, 0], three_attenuate3))
Delay4_1dB = np.column_stack((StereoDelay4[:, 0], three_attenuate4))
Delay5_1dB = np.column_stack((StereoDelay5[:, 0], three_attenuate5))

#Write attenuated by 6dB signals
write('team[Obermaier]-stereosoundfile-[0]-[-6dB].wav', fs, Delay1_6dB)
write('team[Obermaier]-stereosoundfile-[avghead]-[-6dB].wav', fs, Delay2_6dB)
write('team[Obermaier]-stereosoundfile-[1]-[-6dB].wav', fs, Delay3_6dB)
write('team[Obermaier]-stereosoundfile-[10]-[-6dB].wav', fs, Delay4_6dB)
write('team[Obermaier]-stereosoundfile-[100]-[-6dB].wav', fs, Delay5_6dB)

#Write attenuated by 3dB signals
write('team[Obermaier]-stereosoundfile-[0]-[-3dB].wav', fs, Delay1_3dB)
write('team[Obermaier]-stereosoundfile-[avghead]-[-3dB].wav', fs, Delay2_3dB)
write('team[Obermaier]-stereosoundfile-[1]-[-3dB].wav', fs, Delay3_3dB)
write('team[Obermaier]-stereosoundfile-[10]-[-3dB].wav', fs, Delay4_3dB)
write('team[Obermaier]-stereosoundfile-[100]-[-3dB].wav', fs, Delay5_3dB)

#Write attenuated by 1.5dB signals
write('team[Obermaier]-stereosoundfile-[0]-[-1.5dB].wav', fs, Delay1_1dB)
write('team[Obermaier]-stereosoundfile-[avghead]-[-1.5dB].wav', fs, Delay2_1dB)
write('team[Obermaier]-stereosoundfile-[1]-[-1.5dB].wav', fs, Delay3_1dB)
write('team[Obermaier]-stereosoundfile-[10]-[-1.5dB].wav', fs, Delay4_1dB)
write('team[Obermaier]-stereosoundfile-[100]-[-1.5dB].wav', fs, Delay5_1dB)
```
